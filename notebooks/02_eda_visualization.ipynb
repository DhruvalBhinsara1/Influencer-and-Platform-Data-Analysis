### step 2. opening the csv file

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the path works regardless of where the notebook is run from
data_path = os.path.join(os.path.dirname(os.path.abspath("__file__")), '..', 'data', 'cleaned_dataset.csv')
if not os.path.exists(data_path):
    # fallback to direct relative path if running from project root
    data_path = os.path.join('data', 'cleaned_dataset.csv')
df = pd.read_csv(data_path)
df
```

### step 3. Descriptive stats for key numeric columns.

```python
df.describe() 
```

### Task 1. Distribution plots for likes, comments, shares, engagement rate, revenue.

```python
import os
os.makedirs('results', exist_ok=True)
fig, axs = plt.subplots(3, 2, figsize=(8, 8))
axs[0, 0].hist(df['likes'], bins=30, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Distribution of Likes')
axs[0, 0].set_xlabel('Likes')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].hist(df['comments'], bins=30, color='lightgreen', edgecolor='black')
axs[0, 1].set_title('Distribution of Comments')
axs[0, 1].set_xlabel('Comments')
axs[0, 1].set_ylabel('Frequency')

axs[1, 0].hist(df['shares'], bins=30, color='salmon', edgecolor='black')
axs[1, 0].set_title('Distribution of Shares')
axs[1, 0].set_xlabel('Shares')
axs[1, 0].set_ylabel('Frequency')

axs[1, 1].hist(df['engagement_rate'], bins=30, color='plum', edgecolor='black')
axs[1, 1].set_title('Distribution of Engagement Rate')
axs[1, 1].set_xlabel('Engagement Rate')
axs[1, 1].set_ylabel('Frequency')

axs[2, 0].hist(df['revenue_generated'], bins=30, color='gold', edgecolor='black')
axs[2, 0].set_title('Distribution of Revenue')
axs[2, 0].set_xlabel('Revenue')
axs[2, 0].set_ylabel('Frequency')

axs[2, 1].axis('off')
plt.tight_layout()
plt.savefig('results/distribution_plots.png')
plt.show()
```

### Task 2. Correlation heatmap between engagement metrics and revenue/ad_spent.

```python
import os
os.makedirs('results', exist_ok=True)
#Correlation heatmap between engagement metrics and revenue/ad_spent. and true

corr = df[['likes', 'comments', 'shares', 'engagement_rate', 'revenue_generated', 'ad_spent']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.savefig('results/correlation_heatmap.png')
plt.show()
```

### Task 3. Category & platform performance analysis.

```python
import os
os.makedirs('results', exist_ok=True)
#Category & platform performance analysis. with a scatter plot

category_platform = df.groupby(['category', 'platform']).agg({
    'likes': 'mean',
    'comments': 'mean',
    'shares': 'mean',
    'engagement_rate': 'mean',
    'revenue_generated': 'mean',
    'ad_spent': 'mean'
}).reset_index()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=category_platform, x='engagement_rate', y='revenue_generated', hue='category', style='platform', s=100)
plt.title('Category & Platform Performance Analysis')
plt.xlabel('Average Engagement Rate')
plt.ylabel('Average Revenue Generated')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('results/category_platform_performance.png')
plt.show()

print(category_platform)
```

### Task 4. Identify anomalies (e.g., high revenue, low engagement).

```python
import os
os.makedirs('results', exist_ok=True)
anomalies = df[(df['revenue_generated'] > df['revenue_generated'].quantile(0.90)) & (df['engagement_rate'] < df['engagement_rate'].quantile(0.1))]
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='engagement_rate', y='revenue_generated', alpha=0.2)
sns.scatterplot(data=anomalies, x='engagement_rate', y='revenue_generated', color='red', s=50, label='Anomalies')
plt.title('Anomalies: High Revenue, Low Engagement')
plt.xlabel('Engagement Rate')
plt.ylabel('Revenue Generated')
plt.legend()
plt.savefig('results/anomalies_plot.jpg', format='jpeg')
plt.show()

print(anomalies)
```
